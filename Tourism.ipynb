```python
# Load the data from a specific sheet
file_path = "C:/Users/User/Downloads/Domestic_tourism.xlsx"  # Replace with the actual path
sheet_name = 'D4'  # Replace with the actual sheet name

df = pd.read_excel(file_path, sheet_name=sheet_name)

# Pivot the table
pivot_df = df.pivot_table(index=['Year', 'Month', 'Region'], columns='Indicator', values='Value').reset_index()

# Flatten the MultiIndex columns if necessary
pivot_df.columns.name = None  # Remove the name from the columns
pivot_df.columns = [col if isinstance(col, str) else col[1] for col in pivot_df.columns]

# Display the result
print(pivot_df)

output_file_path = 'C:/Users/96659/Downloads/pivoted_data_D4.csv'  # Replace with the desired save location
pivot_df.to_csv(output_file_path, index=False)

print(f"Data saved to {output_file_path}")
```
